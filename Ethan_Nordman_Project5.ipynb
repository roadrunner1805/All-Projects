```r
myDF <- read.csv("/anvil/projects/tdm/data/flights/subset/2006.csv")
```

```r
options(repr.matrix.max.cols=100, repr.matrix.max.rows=100)
head(myDF)
```

```r
pa_flights <- myDF[myDF$Dest == "PHL", ]
```

```r
length(unique(pa_flights$Origin))
```

```r
tapply(pa_flights$AirTime, pa_flights$Origin, mean, na.rm=TRUE)
```

```r
phx_flights <- myDF[myDF$Dest == "PHX", ]
tapply(phx_flights$AirTime, phx_flights$Origin, mean, na.rm=TRUE)
```

```r
# Count flights by destination
dest_counts <- table(myDF$Dest)

# Sort by busiest
sorted_dests <- sort(dest_counts, decreasing=TRUE)

# Top 20 destinations
top_20 <- names(head(sorted_dests, 20))
top_20
```

```r
# names(table(myDF$Dest)) would just give you all destinations in alphabetical order, not in order of busyness
# By sorting before names(), we make sure we only grab the actual busiest airports (top 20 by count), instead of just the first 20 alphabetically.
```

```r
top_dests_df <- myDF[myDF$Dest %in% top_20, ]

tapply(top_dests_df$AirTime, 
       list(top_dests_df$Origin, top_dests_df$Dest), 
       mean, na.rm=TRUE)
```

```r
ice_cream <- read.csv("/anvil/projects/tdm/data/icecream/combined/products.csv")
```

```r
summary(ice_cream$rating)
table(ice_cream$rating)
```

```r
ice_cream$rating_phrases <- cut(
  ice_cream$rating,
  breaks = c(min(ice_cream$rating),
             quantile(ice_cream$rating, 0.25),
             median(ice_cream$rating),
             quantile(ice_cream$rating, 0.75),
             max(ice_cream$rating)),
  labels = c("Wouldn't Recommend",
             "Needs Improvement",
             "Solid Choice",
             "Fan Favorite"),
  include.lowest = TRUE
)
```

```r
head(ice_cream$rating_phrases)
table(ice_cream$rating_phrases)
```

```r
my_flights <- read.csv("/anvil/projects/tdm/data/flights/subset/1997.csv")
```

```r
my_flights$depart_times <- cut(
  my_flights$DepTime,
  breaks = c(0, 300, 600, 900, 1200, 1500, 1800, 2100, 2400),
  labels = c("Night (0-3am)", 
             "Early Morning (3-6am)", 
             "Morning (6-9am)",
             "Late Morning (9-12pm)", 
             "Afternoon (12-3pm)", 
             "Late Afternoon (3-6pm)", 
             "Evening (6-9pm)", 
             "Late Evening (9-12am)"),
  include.lowest = TRUE
)
```

```r
barplot(table(my_flights$depart_times),
        main="Number of Flight Departures by Time of Day (1997)",
        xlab="Time of Day",
        ylab="Number of Flights",
        las=2,
        col="skyblue")
```

```r
table(my_flights$depart_times)
```

```r
barplot(table(my_flights$depart_times),
        main = "Number of Flight Departures by Time of Day (1997)",
        xlab = "Time of Day",
        ylab = "Number of Flights",
        las = 2,      # rotates labels
        col = "skyblue")
```

```r
my_flights$arrival_times <- cut(
  my_flights$ArrTime,
  breaks = c(0, 300, 600, 900, 1200, 1500, 1800, 2100, 2400),
  labels = c("Night (0-3am)", 
             "Early Morning (3-6am)", 
             "Morning (6-9am)",
             "Late Morning (9-12pm)", 
             "Afternoon (12-3pm)", 
             "Late Afternoon (3-6pm)", 
             "Evening (6-9pm)", 
             "Late Evening (9-12am)"),
  include.lowest = TRUE
)
```

```r
table(my_flights$arrival_times)
```

```r
barplot(table(my_flights$arrival_times),
        main = "Number of Flight Arrivals by Time of Day (1997)",
        xlab = "Time of Day",
        ylab = "Number of Flights",
        las = 2,
        col = "lightgreen")
```

```r
# Departures are the heaviest in the mornings and afternoons, making nights and afternoons the least heavy
```

```r
march <- subset(my_flights, Month == 3)
table(march$DayOfWeek)   # check flight counts by weekday in March
```

```r
table(my_flights$DayOfWeek)
```

```r
tapply(my_flights$AirTime, my_flights$DayOfWeek, sum, na.rm=TRUE)
```

```r
flight_times <- tapply(my_flights$AirTime,
                       list(my_flights$DayOfWeek, my_flights$Month),
                       sum,
                       na.rm=TRUE)
```

```r
heatmap(flight_times,
        Rowv = NA, Colv = NA,
        main = "Total AirTime by Day of Week and Month (1997)",
        xlab = "Month",
        ylab = "Day of Week",
        col = heat.colors(256))
```

```r
day_month_times <- tapply(my_flights$AirTime,
                          list(my_flights$DayofMonth, my_flights$Month),
                          sum,
                          na.rm=TRUE)
```

```r
heatmap(day_month_times,
        Rowv = NA, Colv = NA,
        main = "Total AirTime by Day of Month and Month (1997)",
        xlab = "Month",
        ylab = "Day of Month",
        col = heat.colors(256))
```

```r
# Rowv = NA and Colv = NA to avoid hierarchical clustering.
# Adjusted col = heat.colors(256) for readability (you could also try terrain.colors() or topo.colors()).
# Added titles, axis labels to make clear what the rows/columns represent.
```
