# Retrive the data from tables

### Get all of the data from both the Sales and the Department table

```python
sales_info = "SELECT * FROM sales"
sales_df = pd.read_sql_query(sales_info, engine)

dept_info = "SELECT * FROM departments"
dept_df = pd.read_sql_query(dept_info, engine)

merged_df = sales_df.merge(dept_df, on='Dept')

print(merged_df.head())
```

# Show the weekly sales of the Media and Gaming Department

```python
media_gaming_df = merged_df[merged_df['DeptDescription'] == 'Media and Gaming']
print(media_gaming_df)
```

# Show the monthly sales

```python
media_gaming_df['SalesDate'] = pd.to_datetime(media_gaming_df['SalesDate'])

media_gaming_df['Month'] = media_gaming_df['SalesDate'].dt.to_period('M') # thsi is creating the month column

monthly_sales = media_gaming_df.groupby('Month')['Weekly_Sales'].sum()

print(monthly_sales)
```

```python
# Monthly Sales in Millions 

monthly_sales_million = monthly_sales / 1_000_000
print(monthly_sales_million)
```

# Find the expenses for Media and Gaming

```python
query = """
select ea.ExpenseDate, ea.AllocatedAmount, ea.Store, ea.Dept
FROM expense_allocation ea
WHERE ea.Dept = 55
"""

expense_df = pd.read_sql_query(query, engine)

print(expense_df)
```

```python
# Calculate expenses by month

expense_df['ExpenseDate'] = pd.to_datetime(expense_df['ExpenseDate'])

expense_df['Month'] = expense_df['ExpenseDate'].dt.to_period('M')

monthly_expense = expense_df.groupby('Month')['AllocatedAmount'].sum()

monthly_expense_million = monthly_expense / 1_000_000

print(monthly_expense_million)
```

# Caluclate Profit

```python
monthly_profit_million = monthly_sales_million - monthly_expense_million

print(monthly_profit_million)
```

```python
sales_sum = monthly_sales_million.sum()
expense_sum = monthly_expense_million.sum()

print(sales_sum)
print(expense_sum)
```

```python
net_loss = sales_sum - expense_sum

print(net_loss)
```

```python
summary_df = pd.DataFrame({
    'Sales ($M)': monthly_sales_million,
    'Expenses ($M)': monthly_expense_million,
    'Profti/Loss ($M)': monthly_profit_million
})

print(summary_df)

print(f'Total sales: {sales_sum}')
print(f'Total expenses: {expense_sum}')
print(f'Total Profit/Loss: {net_loss}')
```
